```python
import pandas as pd
import numpy as np

df = pandas.read_csv('./combined.csv')
df = df[['date', 'mealcarbs', 'glucose', 'humalog', 'meal']]

#These are required datapoints
breakfast = df[(df['meal'] == 'breakfast') 
               & (np.isfinite(df['glucose'])) 
               & (np.isfinite(df['humalog']))
               & (np.isfinite(df['mealcarbs']))]

lunch = df[(df['meal'] == 'lunch') 
           & (np.isfinite(df['glucose'])) 
           & (np.isfinite(df['humalog']))
           & (np.isfinite(df['mealcarbs']))]

dinner = df[(df['meal'] == 'dinner') 
            & (np.isfinite(df['glucose']))
            & (np.isfinite(df['humalog']))
            & (np.isfinite(df['mealcarbs']))]

#These could be missing
afternoon_snack = df[(df['meal'] == 'afternoon snack') 
            & (np.isfinite(df['mealcarbs']))]

sugar_to_treat = df[(df['meal'] == 'sugar to treat') 
            & (np.isfinite(df['mealcarbs']))]

exercise_snack = df[(df['meal'] == 'exercise snack') 
            & (np.isfinite(df['mealcarbs']))]

#join by date

days_bld = pandas.merge(breakfast, lunch, on='date', suffixes=('_breakfast', ''))
days_bld = pandas.merge(days_bld, dinner, on='date', suffixes=('_lunch', ''))
days_bld = pandas.merge(days_bld, sugar_to_treat, how='left', on='date', suffixes=('_dinner', ''))
days_bld = pandas.merge(days_bld, afternoon_snack, how='left', on='date', suffixes=('_sugar', ''))
days_bld = pandas.merge(days_bld, exercise_snack, how='left', on='date', suffixes=('_snack', '_exercisesnack'))

days_bld[['mealcarbs_snack', 'mealcarbs_sugar', 'mealcarbs_exercisesnack']] = days_bld[['mealcarbs_snack', 'mealcarbs_sugar', 
                                                                                  'mealcarbs_exercisesnack']].fillna(0)
days_bld[['humalog_snack', 'humalog_exercisesnack', 'humalog_sugar']] = days_bld[['humalog_snack', 'humalog_exercisesnack', 'humalog_sugar']].fillna(0)

snacks = pd.DataFrame(days_bld.groupby('date').sum())[['mealcarbs_exercisesnack', 'humalog_exercisesnack', 'mealcarbs_snack',
                                                       'humalog_snack', 'mealcarbs_sugar', 'humalog_sugar']]


snacks['date'] = snacks.index

days_bld = pandas.merge(days_bld, snacks, on='date', how='inner', suffixes=('', '_total'))
days_bld = days_bld.drop_duplicates('date')

max_glucose = pandas.DataFrame(df.groupby('date')['glucose'].max())
max_glucose['date'] = max_glucose.index

days_bld = pandas.merge(days_bld, max_glucose, on='date', how='inner')

```

```python
df_sugar = pandas.read_csv('./combined.csv')
sugar_to_treat = df_sugar[(df_sugar['meal'] == 'sugar to treat') 
            & (np.isfinite(df_sugar['mealcarbs']))]
sugar_to_treat
plt.figure()
p = sns.distplot(sugar_to_treat['hour'])
p.set_title("Sugar to treat by hour of day")
p.set(xlim=(0, 24))

```

```python
days_bld['total_mealcarbs'] = days_bld['mealcarbs_breakfast'] \
                               + days_bld['mealcarbs_breakfast'] \
                               + days_bld['mealcarbs_lunch'] \
                               + days_bld['mealcarbs_dinner'] \
                               + days_bld['mealcarbs_snack'] \
                               + days_bld['mealcarbs_sugar'] \
                               + days_bld['mealcarbs_exercisesnack']

days_bld['total_humalog'] = days_bld['humalog_breakfast'] \
                               + days_bld['humalog_breakfast'] \
                               + days_bld['humalog_lunch'] \
                               + days_bld['humalog_dinner'] \
                               + days_bld['humalog_snack'] \
                               + days_bld['humalog_sugar'] \
                               + days_bld['humalog_exercisesnack']


days_bld['had_exercisesnack'] = days_bld['meal_exercisesnack'].apply(lambda x: not pd.isnull(x))
days_bld['had_sugar_to_treat'] = days_bld['meal_sugar'].apply(lambda x: not pd.isnull(x))
days_bld['had_snack'] = days_bld['meal_snack'].apply(lambda x: not pd.isnull(x))

days_bld['date'] = pandas.to_datetime(days_bld['date'])

next_days_bld = days_bld.copy()
next_days_bld['date'] = pandas.to_datetime(next_days_bld['date'])
next_days_bld['date'] = next_days_bld['date'] - pandas.DateOffset(1)

two_days = days_bld.merge(next_days_bld, on='date', suffixes=('_today', '_tomorrow'))

sns.distplot(two_days['glucose_breakfast_tomorrow'])
#top 15% of measurements are "high glucose"
two_days['glucose_breakfast_tomorrow'].quantile([.25, .5, .80, .85])
#0.85    237.0
two_days['glucose_breakfast_tomorrow_high'] = two_days['glucose_breakfast_tomorrow'] > 237
```

```python
from sklearn.preprocessing import StandardScaler
features = ['glucose_today', 'total_mealcarbs_today',  'total_humalog_today', 
            'had_sugar_to_treat_today','had_exercisesnack_today',
           'humalog_snack_total_today', 'mealcarbs_snack_total_today', 
           'mealcarbs_exercisesnack_total_today', 'mealcarbs_sugar_total_today', 
            'mealcarbs_lunch_today', 'mealcarbs_breakfast_today','mealcarbs_dinner_today', 
            'humalog_breakfast_today', 'humalog_lunch_today', 'humalog_dinner_today', 
            'glucose_breakfast_today', 'glucose_lunch_today', 'glucose_dinner_today']

predict = ['glucose_breakfast_tomorrow']
#foreach of the features scale them
scaler = StandardScaler()
two_days[features] = scaler.fit_transform(two_days[features])
two_days[predict] = scaler.fit_transform(two_days[predict])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(two_days[features], two_days[predict], random_state=30)
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

lr_model = linear_model.LinearRegression()
lr_model.fit(x_train, y_train)
lr_model.score(x_test, y_test)

feature_weights = pandas.DataFrame(list(zip(list(features), list(lr_model.coef_[0])))).rename(index=str, columns={0: 'feature', 1: 'weight'})
feature_weights

plt.figure(figsize=(20, 10))
feature_weights = feature_weights.sort_values(by='weight', ascending=False)
feature_plot = seaborn.barplot(x=feature_weights['feature'], y=feature_weights['weight'])
feature_plot.set_xlabel('Feature', fontsize=20)
feature_plot.set_ylabel('Weight', fontsize=20)
feature_plot.set_xticklabels(feature_plot.get_xticklabels(), rotation=40)

plt.show()

mse_train = mean_squared_error(y_train, lr_model.predict(x_train))
mse = mean_squared_error(y_test, lr_model.predict(x_test))
print('mean squared error lr: ', mse)
print('lr model score: ', lr_model.score(x_test, y_test))

#predictions = pandas.DataFrame.from_dict({'predicted': lr_model.predict(x_test), 'actual': y_test})
predictions = lr_model.predict(x_test).flatten()
actual = y_test

plt.figure(figsize = (30, 30))

df = pandas.DataFrame({'predicted':predictions, 'actual':actual['glucose_breakfast_tomorrow']})

p = seaborn.lmplot(x="predicted", y="actual",
               truncate=True, data=df)
plt.savefig('acutal_predicted_lr_model.png')
plt.show()

print('mse train', mse_train)
print('score: ', r2_score(lr_model.predict(x_train), y_train))
```

```python
import itertools 
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45, ha='right')
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

# Compute confusion matrix
cnf_matrix = confusion_matrix(np.array(gbc_1.predict(x_test)).ravel(), np.array(y_test).ravel()).ravel()

cnf_matrix = np.array([[21 , 0], [5 ,16]])
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
plt.figure(figsize=(10, 10))
plot_confusion_matrix(cnf_matrix, classes=['glucose_in_range_breakfast', 'high_glucose_breakfast'],
                      title='Confusion matrix, without normalization')

# Plot normalized confusion matrix
#plt.figure()
#plot_confusion_matrix(cnf_matrix, classes=class_names, normalize=True,
#                      title='Normalized confusion matrix')

plt.show()
```
